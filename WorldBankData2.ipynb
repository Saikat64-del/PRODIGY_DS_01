```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path1 = 'WorldBank_Data1.xlsx'
file_path2 = 'WorldBank_Data2.xlsx'
df1 = pd.read_excel(file_path1)
df2 = pd.read_excel(file_path2)
```

```python
df1
df2
```

```python
male_pop_cols = [col for col in df1.columns if col.isdigit()]
female_pop_cols = [col for col in df2.columns if col.isdigit()]


print("Missing values in male population columns:")
print(df1[male_pop_cols].isnull().sum())
print("\nMissing values in female population columns:")
print(df2[female_pop_cols].isnull().sum())


print("\nData types of male population columns:")
print(df1[male_pop_cols].dtypes)
print("\nData types of female population columns:")
print(df2[female_pop_cols].dtypes)


print("\nDistribution of male population columns:")
print(df1[male_pop_cols].describe())
print("\nDistribution of female population columns:")
print(df2[female_pop_cols].describe())


if '1960' in male_pop_cols and '1960' in female_pop_cols:
    print("\nCorrelation between male and female populations in 1960:")
    print(df1['1960'].corr(df2['1960']))
```

```python
df = pd.merge(df1, df2, on='Country Name', how='inner', suffixes=('_male', '_female'))

# Select relevant columns
years = [str(year) for year in range(1960, 2024)]
columns_to_keep = ['Country Name'] + [year + '_male' for year in years] + [year + '_female' for year in years]
df = df[columns_to_keep]

# Handle missing values
df.dropna(subset=[col for col in df.columns if col not in ['Country Name']], inplace=True)

# Calculate gender ratio
for year in years:
    df[year + '_ratio'] = df[year + '_female'] / df[year + '_male']

# Select a column for visualization (e.g., the gender ratio for 2023)
visualization_column = '2023_ratio'
df_visualization = df[['Country Name', visualization_column]]
display(df_visualization.head())
```

```python
plt.figure(figsize=(15, 6))
plt.hist(df_visualization['2023_ratio'], bins=20, color='skyblue')  
plt.xlabel('Gender Ratio (Female/Male)')
plt.ylabel('Frequency')
plt.title('Distribution of Gender Ratio in 2023')
plt.tight_layout()
plt.show()
```

```python
top10_countries = df_visualization.sort_values(by='2023_ratio', ascending=False).head(10)
bottom10_countries = df_visualization.sort_values(by='2023_ratio', ascending=True).head(10)
combined_df = pd.concat([top10_countries, bottom10_countries])
plt.figure(figsize=(12, 6))
bar_width = 0.35
predicted_2024 = combined_df['2023_ratio'] * 1.01 
predicted_2025 = combined_df['2023_ratio'] * 1.02
predicted_2026 = combined_df['2023_ratio'] * 1.03
plt.bar(combined_df['Country Name'], combined_df['2023_ratio'], bar_width, label='2023')
plt.bar(combined_df['Country Name'], predicted_2024 - combined_df['2023_ratio'], bar_width, bottom=combined_df['2023_ratio'], label='2024 (Pred)')
plt.bar(combined_df['Country Name'], predicted_2025 - predicted_2024, bar_width, bottom=predicted_2024, label='2025 (Pred)')
plt.bar(combined_df['Country Name'], predicted_2026 - predicted_2025, bar_width, bottom=predicted_2025, label='2026 (Pred)')
plt.xlabel("Country")
plt.ylabel("Gender Ratio (Female/Male)")
plt.title("Gender Ratio in 2023 and Predicted 2024-2026 (Top and Bottom 10)")
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```
