```python
plt.figure(figsize=(14, 7))
data = df['2023']
iqr = np.percentile(data, 75) - np.percentile(data, 25)
bin_width = 2 * iqr / (len(data)**(1/3))
bins = np.arange(min(data), max(data) + bin_width, bin_width)
plt.hist(df['2023'], bins=bins, color='skyblue', edgecolor='black', alpha=0.7)
plt.xscale('log')
plt.xlabel('Population in 2023 (Log Scale)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Population in 2023', fontsize=14)
plt.grid(axis='y', alpha=0.75, linestyle='--')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.axvline(df['2023'].mean(), color='red', linestyle='dashed', linewidth=1.5, label=f'Mean: {df["2023"].mean():.2e}')
plt.axvline(df['2023'].median(), color='green', linestyle='dashed', linewidth=1.5, label=f'Median: {df["2023"].median():.2e}')
plt.legend()
plt.show()
```
